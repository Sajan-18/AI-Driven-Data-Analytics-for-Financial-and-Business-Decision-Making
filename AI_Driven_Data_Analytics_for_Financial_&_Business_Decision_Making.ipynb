```python
total_sales = round(df["sales"].sum(),2)

total_orders = df["order_id"].nunique()

total_customers = df["customer_id"].nunique()

total_products = df["product_id"].nunique()

average_sales = round(df["sales"].mean(),2)

print("========== BUSINESS KPIs ==========")

print("Total Sales :", total_sales)

print("Total Orders :", total_orders)

print("Total Customers :", total_customers)

print("Total Products :", total_products)

print("Average Sales :", average_sales)
```

```python
category_sales = (
    df.groupby("category")["sales"]
      .sum()
      .sort_values(ascending=False)
)

print(category_sales)

best_category = category_sales.idxmax()

print()

print("Best Performing Category :", best_category)
```

```python
region_sales = (
    df.groupby("region")["sales"]
      .sum()
      .sort_values(ascending=False)
)

print(region_sales)

best_region = region_sales.idxmax()

print()

print("Best Region :", best_region)
```

```python
state_sales = (
    df.groupby("state")["sales"]
      .sum()
      .sort_values(ascending=False)
)

print(state_sales.head(10))
```

```python
customer_sales = (
    df.groupby("customer_name")["sales"]
      .sum()
      .sort_values(ascending=False)
)

print(customer_sales.head(10))
```

```python
product_sales = (
    df.groupby("product_name")["sales"]
      .sum()
      .sort_values(ascending=False)
)

print(product_sales.head(10))
```

```python
monthly_sales = (
    df.groupby(df["order_date"].dt.to_period("M"))["sales"]
      .sum()
)

monthly_sales.index = monthly_sales.index.astype(str)

print(monthly_sales)
```

```python
segment_sales = (
    df.groupby("segment")["sales"]
      .sum()
      .sort_values(ascending=False)
)

print(segment_sales)
```

```python
ship_sales = (
    df.groupby("ship_mode")["sales"]
      .sum()
      .sort_values(ascending=False)
)

print(ship_sales)
```

```python
top_state = state_sales.idxmax()

top_customer = customer_sales.idxmax()

top_product = product_sales.idxmax()

best_segment = segment_sales.idxmax()

best_ship_mode = ship_sales.idxmax()

business_metrics = pd.DataFrame({

    "Metric":[

        "Total Sales",
        "Total Orders",
        "Total Customers",
        "Total Products",
        "Average Sales",
        "Best Category",
        "Best Region",
        "Top State",
        "Top Customer",
        "Top Product",
        "Best Customer Segment",
        "Most Preferred Shipping Mode"

    ],

    "Value":[

        total_sales,
        total_orders,
        total_customers,
        total_products,
        average_sales,
        best_category,
        best_region,
        top_state,
        top_customer,
        top_product,
        best_segment,
        best_ship_mode

    ]

})

business_metrics.to_csv(

    os.path.join(REPORTS_DIR, "business_metrics.csv"),

    index=False

)

print("Business Metrics Saved Successfully.")

business_metrics
```
